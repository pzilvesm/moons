```python
#Program to visualize Jupiter's moons and its orbits in scale with a map
#by Giulya Souza and Pedro Zilves

import folium
import pandas as pd
import numpy as np
#from google.colab import files
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/pzilvesm/test/master/moonsjupyter.csv')
Semiax=data.SemMajorAx.to_list()
```

```python
ScaledSemiax=[]
scale=2 # here we define a scale 
for i in Semiax:
  ScaledSemiax.append(i/scale)
Juprad=69911/scale
Diametros=data.Diameter.to_list()
Raios=[]
for i in Diametros: Raios.append(i/2/scale)  #and the radius of the principal moons
Names=data.Name.to_list()
```

```python
print(ScaledSemiax)
```

```python
m = folium.Map(location=[-22.857972080450708, -43.231172583249],zoom_start=8,control_scale=True) #generating a map with the location in coordinates lat, lon
folium.Circle(
    location=[-22.857972080450708, -43.231172583249],
    radius=Juprad,
    popup="Jupiter",
    color="darkkhaki",
    fill=True,
    #fill_color="darkkhaki",
).add_to(m)

for i in ScaledSemiax:
#i=ScaledSemiax[-1]
  folium.Circle( #this function generate a circle in the map 
    location=[-22.857972080450708, -43.231172583249],
    radius=i, #with radius in meters
    #popup="Jupiter",
    color="gray",
    fill=False,
    #fill_color="darkkhaki",
).add_to(m) #adding all the orbital curves of the moons

#now adding the galilean moons to the map
folium.Circle(
    location=[-15.795383636713867, -48.19634445012412],
    radius=Raios[7],
    popup=Names[7],
    color="turquoise",
    fill=True,
    fill_color="turquoise",
).add_to(m)

folium.Circle(
    location=[-20.752183909763048, -47.8928912132001],
    radius=Raios[6],
    popup=Names[6],
    color="lightcoral",
    fill=True,
    fill_color="lightcoral",
).add_to(m)

folium.Circle(
    location=[-19.911097878409, -43.93988250039992],
    radius=Raios[5],
    popup=Names[5],
    color="yellowgreen",
    fill=True,
    fill_color="yellowgreen",
).add_to(m)

folium.Circle(
    location=[-21.128626164259565, -42.38310418897777],
    radius=Raios[4],
    popup=Names[4],
    color="gold",
    fill=True,
    fill_color="gold",
).add_to(m)
```

```python
m.save('mapJupiter.html',close_file=True) #saving and visualizing
#files.download('mapJupiter.html')
m
```
