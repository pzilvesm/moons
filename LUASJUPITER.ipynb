```python
namelist=data['Name'].to_list()

plt.figure(figsize=(19,4))

X=np.arange(len(namelist))

plt.style.use('ggplot') #style of the plots
container=plt.bar(X,data['SemMajorAx'],label="Semi-major axis") #plotting in bars
plt.ylabel("Semi-major axis(km)")
plt.xticks(ticks=X)
#plt.bar_label(container,labels=namelis[: i])
#plt.savefig('teste.png')
plt.show()
```

```python
Diametros=data.Diameter.to_list()
Raios=[]
for i in sorted(Diametros,reverse=True): Raios.append(i/2) # putting into a ordered list
```

```python
colors = itertools.cycle(["r", "b", "g"]) #loop to choose the color of the moons with larger Semimajor axes

j=0 # j is the distance of the circles in the figure

plt.figure(figsize=(15,8.5))
for i in Raios:
 
  if i>1000:
    j+=3000
  elif i>80:
    j+=2000
  elif i>10:
    j+=500
  else: j+=10

  if i>2600: 
    plt.gca().add_patch(Circle((j,0),i,color="lightcoral",label='Ganymede'))
  elif i>2000: 
    plt.gca().add_patch(Circle((j,0),i,color="turquoise",label='Callisto'))
  elif i>1800: 
    plt.gca().add_patch(Circle((j,0),i,color="gold",label='Io'))
  elif i>1500: 
    plt.gca().add_patch(Circle((j,0),i,color="yellowgreen",label='Europa'))
  else: 
    plt.gca().add_patch(Circle((j,0),i,color=next(colors)))

plt.gca().add_patch(Circle((10000,73911),69911,color="darkkhaki",label='Jupiter'))


plt.xlim([-200,20000])#
plt.ylim([-3000,8000])#
plt.xlabel("[Km]")
plt.ylabel("[Km]")
plt.title('Size of Jupite$\'$s Moons ')
plt.legend()
plt.show()
```
